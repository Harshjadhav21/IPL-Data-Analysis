# 3.Visulizing the data

# finding top teams and players

```python
#toss win by count
toss_win = ipl_df.toss_winner.value_counts()
toss_win
```

```python
toss_win[toss_win >= 10 ].plot(kind = 'bar', title = 'Toss win count by team',figsize=(15,4))
```

```python
#top 10 players of match
ipl_df = ipl_df.head(10).copy()
ipl_df.rename(columns={'id':'Total_Awards','player_of_match':'Man_Of_The_Match'},inplace=True)
ipl_df
```

# Finding the factpor affecting the victory

```python
#Teams that won the toss
sns.countplot(x='toss_winner', data=ipl_df).set(title='Teams that won the toss')
plt.xticks(rotation=90)
plt.show()
```

# Data analysis:

# No of matches played in each stadium

```python
total_venue = ipl_df.venue.value_counts()
total_venue
```

```python
total_venue[total_venue >= 10 ].plot(kind = 'bar', title = 'Total match played in each venue',figsize=(15,4))
```

```python
#team that won the matches by batting first
batting_first=ipl_df[ipl_df['win_by_runs']!=0]
batting_first['winner'].value_counts()
```

```python
#teams who had won the toss and won the match
win = ipl_df.winner == ipl_df.toss_winner
win_counts = win.value_counts()
win_counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Does winning the toss increase the chances of winning the match?')
plt.pie(win_counts,labels=win_counts.index,autopct='%1.1f%%', startangle=180);
```

# Batting stats:

```python
#most runs score by team
deliveries_df.groupby(['batting_team'])['total_runs'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,8))
data = deliveries_df.groupby(['batting_team'])['total_runs'].sum().sort_values(ascending=False)
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('batting_team')
plt.ylabel('total_runs')
plt.show()
```

```python
# Most IPL Runs by a Batsman
plt.figure(figsize=(10,8))
data = deliveries_df.groupby(['batsman'])['batsman_runs'].sum().sort_values(ascending=False)[:5]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Batsman')
plt.ylabel('batsman_Runs')
plt.show()
```

```python
#total runs
data = deliveries_df.groupby(['batsman'])['batsman_runs'].sum().sort_values(ascending=False)[:5]
data
```

# bowling stats:

```python
#extra runs by bowler
plt.figure(figsize=(10,8))
data = deliveries_df.groupby(['bowler'])['extra_runs'].sum().sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('bowler')
plt.ylabel('extra_runs')
plt.show()
```

```python
data = deliveries_df.groupby(['bowler'])['extra_runs'].sum().sort_values(ascending=False)[:10]
data
```

```python
# total no balls by bowler
plt.figure(figsize=(10,8))
data = deliveries_df.groupby(['bowler'])['noball_runs'].sum().sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('bowler')
plt.ylabel('noball_runs')
plt.show()
```

```python
data = deliveries_df.groupby(['bowler'])['noball_runs'].sum().sort_values(ascending=False)[:10]
data
```

```python
#extra runs by teams
plt.figure(figsize=(10,8))
data = deliveries_df.groupby(['bowling_team'])['extra_runs'].sum().sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('bowling_team')
plt.ylabel('extra_runs')
plt.show()
```

```python
data = deliveries_df.groupby(['bowling_team'])['extra_runs'].sum().sort_values(ascending=False)[:10]
data
```
